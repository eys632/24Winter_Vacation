```python
import re

# 텍스트 파일 읽기
file_path = '/content/appendix-keywords.txt'
with open(file_path, 'r', encoding='utf-8') as file:
    file_content = file.read()

# 정규식을 사용하여 텍스트를 묶음 단위로 분리
pattern = r"(.*?\n정의:.*?\n예시:.*?\n연관키워드:.*?(?:\n|$))"
blocks = re.findall(pattern, file_content, re.DOTALL)

# 각 묶음을 출력
for i, block in enumerate(blocks):
    block = block.strip()  # 공백 제거
    print(block)
    if i < len(blocks) - 1:  # 마지막 블록 이후에는 구분선 출력하지 않음
        print("=" * 80)
```

```python
from langchain.text_splitter import TokenTextSplitter

# 텍스트 파일 읽기
file_path = '/content/appendix-keywords.txt'

with open(file_path, 'r', encoding='utf-8') as file:
    text = file.read()

# TokenTextSplitter를 사용한 텍스트 분할
text_splitter = TokenTextSplitter(chunk_size=100, chunk_overlap=20)  # 링크 내용을 기반으로 적절한 값 사용
chunks = text_splitter.split_text(text)

# 결과 출력
for i, chunk in enumerate(chunks, start=1):
    print(f"청크 {i}:\n{chunk.strip()}")
    print("======")
```
